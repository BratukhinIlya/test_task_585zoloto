# Расчет средней наценки
Необходимо расчитать наценку на каждый товар, проданный с 01.04.2023 по 10.04.2023, а затем расчитать среднюю наценку на товары из серебра и из золота

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('test_task_one.csv', parse_dates=True)
df.head()
```

Мы считали **датафрэйм** из файла

```python
df = df.rename(columns = {'ДатаПродажи' : 'date', 'НомерЧека' : 'check_number', 'Номенклатура' : 'nomenclature', 'Штрихкод' : 'barcode', 
                         'ТоварнаяГруппа' : 'product_group', 'СН_Характеристика' : 'ch_characteristic', 'бгсВес' : 'mass', 
                          'ОбщаяСумма' : 'all_cost', 'Себестоимость' : 'cost_price'})
df.head()
```

Переименовали колонки для удобсва дальнейшей работы с датафрэймом с помощью функции **rename**

```python
right_df = df[df['date'].between('2023-04-01', '2023-04-11')]
right_df.head()
```

Далее извлекаем только те покупки, которые были совершены с **01.04.2023 00:00 по 10.04.2023 23:59**. Что извлекли - представлено выше

```python
right_df = right_df.copy()
right_df['all_cost'] = right_df['all_cost'].str.replace(',', '.', regex=False)
right_df['cost_price'] = right_df['cost_price'].str.replace(',', '.', regex=False)
```

```python
right_df.all_cost = right_df['all_cost'].astype(float)
right_df.cost_price = right_df['cost_price'].astype(float)
right_df_not_zero = right_df.query("all_cost > 0 and cost_price > 0")
```

```python
right_df
```

Далее нам надо было преобразовать колонки **all_cost** и **cost_price** в тип данных с плавающей запятой, чтобы найти все значения больше нуля и создать новую колонку - **наценка (under_price)**. Для этого мы сначала заменяем запятую на точку, потом уже приводим к типу данных **float**. Создаем новую колонку **наценка (under_price)** разницей **общей суммы (all_cost)** и **себестоимостью (cost_price)**

```python
right_df['under_price'] = right_df.all_cost - right_df.cost_price
right_df.head()
```

```python
right_df['product_group'] = right_df['product_group'].astype(str)
right_df['ch_characteristic'] = right_df['ch_characteristic'].astype(str)
silver_df = right_df[right_df['product_group'].str.startswith("СИ")]
gold_df = right_df[right_df['ch_characteristic'].fillna("").str.startswith("Au")]
silver_df.head()
```

Мы приводим к строчному типу данных колонки **product_group** и **ch_characteristic**. И создаем 2 новых датафрэйма с серебряными изделиями и золотыми изделиями, после чего мы находим **средние значения наценки** на каждое изделие

```python
gold_df.head()
```

```python
print(f'Средняя наценка на изделия из золота = {np.round(gold_df['under_price'].mean(), 2)}')
print(f'Средняя наценка на изделия из серебра = {np.round(silver_df['under_price'].mean(), 2)}')
```
